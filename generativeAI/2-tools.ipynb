```python
import os
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model=init_chat_model("groq:llama-3.1-8b-instant")
response=model.invoke("Write an essay on AI in education")
response
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """Get the current weather in a given location."""
    # In a real implementation, you would call a weather API here.
    return f"The current weather in {location} is sunny with a high of 25°C."

model_with_tools=model.bind_tools([get_weather]) 
```

```python
response=model_with_tools.invoke("What is the weather in Guwahati?")
print(response)
```

```python
response.tool_calls
```

### Tools execution

```python
# Step 1 : Model generate tool calls
messages = [{"role": "user", "content": "What is the weather in Guwahati?"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)
print(ai_msg.tool_calls)
print(ai_msg)
print(messages)
print("\n\n")
# STep 2 : Execute tools and collect results
for tool_call in ai_msg.tool_calls:
    # Execute the toll with thye generated arguments
    tool_result = get_weather.invoke(tool_call)
    print(tool_result)
    messages.append(tool_result)
    print(messages)

print("\n\n")
# Step 3 : Pass result back to model for final response
final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
